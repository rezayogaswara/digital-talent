# LATIHAN 2 - Tugas List Nilai

Diberikan sekumpulan Nilai Random antara 0 sampai dengan 100 sebanyak 250 nilai. Buatlah program untuk melakukan hal berikut:
1. Tampilkan Angka Ganjil dari ke 250 nilai tersebut
2. Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
3. Tampilkan standar deviasi dari sekumpulan nilai tersebut
4. Tampilkan nilai maksimum dan minimum
5. Tampilkan modus dari sekumpulan nilai tersebut

```python
import random

# GENERATE RANDOM
nilaiTugas = []

for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print(nilaiTugas)

# TAMPILKAN:
# ganjil
nilaiGanjil = []

for g in nilaiTugas:
    if g % 2 != 0 and not(g in nilaiGanjil) :
        nilaiGanjil.append(g)

print()
print('Angka Ganjil dari ke 250 nilai di atas (angka yang sama hanya ditampilkan satu kali ) yaitu :', nilaiGanjil)

# nilai > 70
nilai70Plus = 0

for b in nilaiTugas:
    if b > 70 :
        nilai70Plus += 1

print()
print('Yang mendapatkan nilai lebih dari 70 berjumlah', nilai70Plus)

# standar deviasi
totalNilai = 0

for n in nilaiTugas:
    totalNilai += n

mean = totalNilai / 250

sigma = 0

for n in nilaiTugas:
    sigma += (n - mean) ** 2
    
varian = sigma / (250 - 1)

standarDeviasi = varian ** 0.5

print()
print('Standar deviasi dari sekumpulan nilai di atas adalah', standarDeviasi)

# maksimum
nilaiMaksimum = 1

for m in nilaiTugas:
    if m > nilaiMaksimum :
        nilaiMaksimum = m

print()
print('Nilai maksimum =', nilaiMaksimum)

# minimum
nilaiMinimum = 99

for m in nilaiTugas:
    if m < nilaiMinimum :
        nilaiMinimum = m

print()
print('Nilai minimum =', nilaiMinimum)

# modus
frekuensiNilai = []

for i in range(0,100) :
    frekuensiNilai.append(0)

for n in nilaiTugas :
    frekuensiNilai[n] += 1

frekuensiModus = 0

for f in frekuensiNilai :
    if f > frekuensiModus :
        frekuensiModus = f

modus = []

for i,f in enumerate(frekuensiNilai) :
    if f == frekuensiModus :
        modus.append(i)

print()
print('Modus dari sekumpulan nilai di atas adalah', modus)
```
