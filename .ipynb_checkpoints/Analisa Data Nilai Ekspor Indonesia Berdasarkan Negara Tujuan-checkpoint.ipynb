# 1. Deskripsi Data
Dataset ini berisi mengenai nilai ekspor Indonesia berdasarkan negara tujuan

Penjelasan mengenai Variabel pada Dataset ini:

1. tahun : Tahun
2. bulan : Bulan
3. benua : Nama Benua
4. region : Nama Region
5. nama_negara : Nama Negara tujuan
6. nilai_ekspor : Nilai ekspor (dalam Ribu USD)

Catatan:

1. Pemekaran menjadi 27 negara sejak Bulgaria dan Romania bergabung dengan Uni Eropa pada Januari 2007
2. Untuk "Ekspor yang tidak dapat diklasifikasikan" terdiri dari barang yang diperoleh di pelabuhan oleh sarana pengangkut, barang untuk diperbaiki, dan penyesuaian cakupan ekspor industri Batam

```python
import pandas as pd

data_ekspor = pd.read_csv('processed-nilai-ekspor-negara-tujuan-2005-2013.csv', delimiter = ',', encoding = "ISO-8859-1")
```

# 2. Statistik & Visualisasi Data Sederhana
## 2.1. Total Nilai Ekspor
### 2.1.1. Nilai Ekspor Indonesia Per Tahun (dalam Ribu USD)

```python
data_ekspor.groupby('tahun').sum()
```

```python
ax = data_ekspor.groupby('tahun').sum()
ax = ax.plot(kind='bar', title ="Nilai Ekspor Indonesia", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai Ekspor (dalam Ribu USD)", fontsize=12)
```

### 2.1.2. Total Nilai Ekspor Indonesia Tahun 2005 - 2013

```python
print(data_ekspor.nilai_ekspor.sum(), 'ribu USD')
```

### 2.1.3. Total Nilai Ekspor Indonesia Per Benua Tahun 2005 - 2013 (dalam Ribu USD)

```python
data_ekspor.groupby('benua').sum()['nilai_ekspor']
```

```python
ax = data_ekspor.groupby('benua').sum()['nilai_ekspor']
ax = ax.plot(kind='bar', title ="Nilai Ekspor Indonesia Per Benua Tahun 2005 - 2013", legend=True, fontsize=12)
ax.set_xlabel("Benua", fontsize=12)
ax.set_ylabel("Nilai Ekspor (dalam Ribu USD)", fontsize=12)
```

### 2.1.4. Total Nilai Ekspor Indonesia Per Region Tahun 2005 - 2013 (dalam Ribu USD)

```python
data_ekspor.groupby(['benua', 'region']).sum()['nilai_ekspor']
```

```python
ax = data_ekspor.groupby('region').sum()['nilai_ekspor']
ax = ax.plot(kind='bar', title ="Nilai Ekspor Indonesia Per Region Tahun 2005 - 2013", legend=True, fontsize=12)
ax.set_xlabel("Region", fontsize=12)
ax.set_ylabel("Nilai Ekspor (dalam Ribu USD)", fontsize=12)
```

### 2.1.5. Total Nilai Ekspor Indonesia Per Negara (dalam Ribu USD)

```python
data_ekspor.groupby('nama_negara').sum()['nilai_ekspor']
```

```python
ax = data_ekspor.groupby('nama_negara').sum()['nilai_ekspor']
ax = ax.plot(kind='bar', title ="Nilai Ekspor Indonesia Per Negara Tahun 2005 - 2013", legend=True, fontsize=12)
ax.set_xlabel("Negara", fontsize=12)
ax.set_ylabel("Nilai Ekspor (dalam Ribu USD)", fontsize=12)
```

## 2.3. Rata - rata Nilai Ekspor

# 3. Narasi Analisis
1. Dalam rentang tahun 2009 - 2015, nilai ekspor Indonesia berada pada tren yang meningkat dari tahun ke tahun. Dari rentang 9 tahun, penurunan hanya terjadi di 3 tahun, yakni tahun 2009, 2012, dan 2013. Perlu penelitian lebih lanjut agar dapat diketahui faktor - faktor apa saja yang berpengaruh terhadap menurunnya nilai ekspor pada 3 tahun tersebut sehingga ke depannya faktor - faktor penghambat tersebut dapat dieliminir.
2. Dalam lingkup benua, benua Asia & Timur Tengah merupakan pasar ekspor terbesar Indonesia dengan nilai ekspor dalam rentang tahun 2005 - 2013 sebesar 912.358.600.000 USD. Penyumbang terbesar dari nilai ini adalah negara - negara di kawasan Asia Timur dan penyumbang terbesar di kawasan ini adalah negara Jepang. Dari data ini, dapat kita simpulkan bahwa kegiatan ekspor ke negara - negara di benua Asia & Timur Tengah, khususnya kawasan Asia Timur, dan tentunya negara Jepang (sebagai penyumbang nilai ekspor terbesar di kawasan ini) merupakan salah satu andalan dalam menambah devisa negara kita yang harus terus dipertahankan di tahun - tahun selanjutnya. Sementara itu, nilai ekspor terkecil dalam rentang tahun 2005 - 2013 dihasilkan dari kegiatan ekspor ke benua Afrika dengan nilai 24.840.890 USD. Nilai ekspor ke benua - benua lain yang belum sebesar nilai ekspor ke benua Asia & Timur Tengah tentu harus dipandang sebagai sebuah potensi pasar yang belum tergarap sehingga para eksportir Indonesia harus mencari cara agar produk Indonesia bisa masuk ke negara - negara yang saat ini belum banyak atau bahkan sama sekali mengimpor produk - produk Indonesia.
3. Jepang terbukti merupakan salah satu pasar ekspor terbaik Indonesia karena dalam rentang tahun 2005 - 2013, rata - rata nilai ekspor Indonesia ke Jepang berada di atas rata - rata nilai ekspor Indonesia ke kawasan Asia Timur, benua Asia & Timur Tengah, dan rata - rata nilai ekspor Indonesia secara keseluruhan. Nilai ekspor tertinggi dalam rentang waktu yang sama ternyata juga adalah dengan tujuan negara ini, yakni selama bulan Mei 2011, Indonesia berhasil melakukan ekspor ke Jepang hingga mencapai nilai 3.068.508 USD.
